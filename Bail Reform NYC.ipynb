```python
import pandas as pd
```

```python
# import url directly only show 1000 rows 
# please download it from NYC open data portal
file_path1 = '/Users/yokij/Desktop/NYPD_Arrests_Data__Historic__20231116.csv'
data1 = pd.read_csv(file_path1)
data1 = data1.drop(columns=['Lon_Lat'])
file_path2 = '/Users/yokij/Desktop/NYPD_Arrest_Data__Year_to_Date__20231116.csv'
data2 = pd.read_csv(file_path2)
data2 = data2.drop(columns=['New Georeferenced Column'])
df = pd.concat([data1, data2])
df.head()
```

```python
#delete data if its category is NaN
df = df[df['KY_CD'].notna()] #KY_CD is more general than PD_CD
df = df[df['LAW_CAT_CD'].notna()] #This records Three categories(felony,misdemeanor and violation)
# Filter the DataFrame to keep records from 2014 to 2023
df['ARREST_DATE'] = pd.to_datetime(df['ARREST_DATE'])
df = df[(df['ARREST_DATE'].dt.year >= 2014) & (df['ARREST_DATE'].dt.year <= 2023)]
df = df.sort_values(by='ARREST_DATE')
```

```python
#Check and Display rows with missing values
rows_with_missing_values = df.isnull().any(axis=1)
missing_value_records = df[rows_with_missing_values]
print(missing_value_records)
```
