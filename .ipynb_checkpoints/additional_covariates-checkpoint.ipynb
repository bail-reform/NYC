# NYC

## P9 Race

Decennial Census 2020: https://data.census.gov/table/DECENNIALDHC2020.P9?q=race&g=040XX00US36$8600000

```python
import pandas as pd
file_path = '/Users/yujinlee/Google_Drive/NYU/Fall_2023/Capstone/CrimeData'
df = pd.read_csv(file_path + '/control/nyc/DECENNIALDHC2020.P9-Data.csv', low_memory=False)
df.head()
```

```python
black = df[['NAME', 'P9_001N', 'P9_006N']]
black = black.iloc[1:].reset_index(drop=True)
black.columns = ['ZCTA5CE10', 'total_pop', 'black']
black['ZCTA5CE10'] = black['ZCTA5CE10'].str.replace('ZCTA5 ', '')
black.head()
```

```python
black['total_pop'] = pd.to_numeric(black['total_pop'], errors='coerce')
black['black'] = pd.to_numeric(black['black'], errors='coerce')

# Create a new column 'percent_black'
black['percent_black'] = black['black'] / black['total_pop']

# Drop 'total_pop' and 'black' columns
black.drop(['total_pop', 'black'], axis=1, inplace=True)

black.head()
```

```python
black.sort_values('percent_black', ascending=False)
```

```python
black = black.fillna(0)
```

```python
black.sort_values('percent_black', ascending=False)
```

```python
black['ZCTA5CE10'] = pd.to_numeric(black['ZCTA5CE10'], errors='coerce')
```

## Merge with Existing

```python
# import data
crime_nyc = pd.read_csv(file_path + '/clean/nyc_weekly_data.csv', low_memory=False)
crime_nyc.head()
```

```python
crime_nyc.dtypes
```

```python
crime_nyc = crime_nyc.merge(black, on = 'ZCTA5CE10', how = 'left')
crime_nyc.head()
```

```python
crime_nyc.to_csv(file_path + '/clean/nyc_weekly_data_withblack.csv')
```

# Boston

## P9 Race

```python
import pandas as pd
df = pd.read_csv(file_path + '/control/boston/DECENNIALDHC2020.P9-Data.csv', low_memory=False)
df.head()
```

```python
df = df.transpose()
df = df.reset_index()
df.head()
```

```python
black = df.iloc[:, [0,1,6]]
black.columns = ['ZCTA5CE10', 'total_pop', 'black']
black = black.iloc[1:].reset_index(drop=True)
black['ZCTA5CE10'] = black['ZCTA5CE10'].str.replace('ZCTA5 ', '')

black.head()
```

```python
black['total_pop'] = black['total_pop'].str.replace(',', '', regex=False)
black['black'] = black['black'].str.replace(',', '', regex=False)
black.head()
```

```python
black['total_pop'] = pd.to_numeric(black['total_pop'], errors='coerce')
black['black'] = pd.to_numeric(black['black'], errors='coerce')
```

```python
# Create a new column 'percent_black'
black['percent_black'] = black['black'] / black['total_pop']

# Drop 'total_pop' and 'black' columns
black.drop(['total_pop', 'black'], axis=1, inplace=True)

black.head()
```

```python
black = black.fillna(0)
```

```python
black.sort_values('percent_black', ascending=False)
```

## Merge with Existing

```python
# import data
crime_boston = pd.read_csv(file_path + '/clean/boston_weekly_data_with_covariates.csv', low_memory=False)
crime_boston.head()
crime_boston['ZCTA5CE10'] = crime_boston['ZCTA5CE10'].astype(str).str.zfill(5)
```

```python
crime_boston = crime_boston.merge(black, on = 'ZCTA5CE10', how = 'left')
crime_boston.head()
```

```python
crime_boston.to_csv(file_path + '/clean/boston_weekly_data_withblack.csv')
```
