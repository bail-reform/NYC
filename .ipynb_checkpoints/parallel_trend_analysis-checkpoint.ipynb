```python
import pandas as pd
file_path = '/Users/yujinlee/Google_Drive/NYU/Fall_2023/Capstone/CrimeData'
```

```python
boston = pd.read_csv(file_path + '/clean/boston_weekly_data_withblack.csv', low_memory = False)
seattle = pd.read_csv(file_path + '/clean/seattle_weekly_data_withblack.csv', low_memory = False)
nyc = pd.read_csv(file_path + '/clean/nyc_weekly_data_withblack.csv', low_memory = False)
la = pd.read_csv(file_path + '/clean/la_weekly_data_withblack.csv', low_memory = False)
```

```python
nyc.head()
```

```python
la = la.rename(columns={"new_case": "case_count"})
boston = boston.rename(columns={"weekly_cases": "case_count"})
```

```python
nyc = nyc.drop(['CASERATE', 'PCTPOS', 'y_hat'], axis = 1)
la = la.drop(['caserate', 'y_hat'], axis=1)
```

```python
la.dtypes
```

```python
nyc[nyc.isna().any(axis=1)]
```

```python
la[la.isna().any(axis=1)]
```

```python
boston['week'] = pd.to_datetime(boston['week'])
seattle['week'] = pd.to_datetime(seattle['week'])
nyc['week'] = pd.to_datetime(nyc['week'])
la['week'] = pd.to_datetime(la['week'])
```

# 1st Pre-trend Analysis

```python
boston_1st = boston[(boston['week'] >= '2018-01-01') & (boston['week'] <= '2019-12-30')]
seattle_1st = seattle[(seattle['week'] >= '2018-01-01') & (seattle['week'] <= '2019-12-30')]
nyc_1st = nyc[(nyc['week'] >= '2018-01-01') & (nyc['week'] <= '2020-04-30')]
la_1st = la[(la['week'] >= '2018-01-01') & (la['week'] <= '2020-04-30')]
```

```python
nyc_1st['city'] = 'nyc'
la_1st['city'] = 'la'
boston_1st['city'] = 'boston'
seattle_1st['city'] = 'seattle'
```

```python
nycla_1st = pd.concat([nyc_1st, la_1st], axis=0)
nycboston_1st = pd.concat([nyc_1st, boston_1st], axis=0)
nycseattle_1st = pd.concat([nyc_1st, seattle_1st], axis=0)
```

```python
nycla_1st
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

nycla_model = smf.ols('crime_rate ~ C(city) + week_no:C(city) + week_no + p_rate + un_rate + in_median + case_count + percent_black', data=nycla_1st).fit(cov_type='cluster', cov_kwds={'groups': nycla_1st['ZCTA5CE10']})
nycla_1st['y_hat_nycla'] = nycla_model.predict(nycla_1st)
print(nycla_1st['y_hat_nycla'].mean())
print(nycla_model.summary())
```

```python
nycboston_1st.head()
```

```python
nycboston_model = smf.ols('crime_rate ~ C(city) + week_no:C(city) + week_no + p_rate + un_rate + in_median + case_count + percent_black', data=nycboston_1st).fit(cov_type='cluster', cov_kwds={'groups': nycboston_1st['ZCTA5CE10']})
nycboston_1st['y_hat_nycboston'] = nycboston_model.predict(nycboston_1st)
print(nycboston_1st['y_hat_nycboston'].mean())
print(nycboston_model.summary())
```

```python
nycseattle_model = smf.ols('crime_rate ~ C(city) + week_no:C(city) + week_no + p_rate + un_rate + in_median + case_count + percent_black', data=nycseattle_1st).fit(cov_type='cluster', cov_kwds={'groups': nycseattle_1st['ZCTA5CE10']})
nycseattle_1st['y_hat_nycseattle'] = nycseattle_model.predict(nycseattle_1st)
print(nycseattle_1st['y_hat_nycseattle'].mean())
print(nycseattle_model.summary())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of weeks
x = np.arange(0, 105)  # From 0 to 130 weeks

# Calculate the crime rate based on the first given equation
y1 = -0.0122 * x

# Calculate the crime rate based on the second given equation
y2 = -0.0119 * x

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the first equation
plt.plot(x, y1, label='Crime Rate = -0.0122 * Weeks')

# Plot the second equation
plt.plot(x, y2, label='Crime Rate = -0.0119 * Weeks')  # Dashed line for distinction

# Labeling the axes
plt.xlabel('Weeks')
plt.ylabel('Crime Rate')
plt.title('Crime Rate Over Time')

# Adding a legend
plt.legend()

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```
