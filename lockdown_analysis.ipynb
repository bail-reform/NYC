### Boston Data

```python
import pandas as pd

file_path = '/Users/yujinlee/Google_Drive/NYU/Fall_2023/Capstone/CrimeData'
df = pd.read_csv(file_path + '/boston_weekly_data.csv', low_memory = False)
df.head()
```

```python
df['occurred_on_date'] = pd.to_datetime(df['occurred_on_date'])
```

```python
# lockdown 2020
df[(df['occurred_on_date'] >= pd.Timestamp('2020-03-15')) & (df['occurred_on_date'] <= pd.Timestamp('2020-07-01'))]
```

```python
# lockdown 2018
df[(df['occurred_on_date'] >= pd.Timestamp('2018-03-15')) & (df['occurred_on_date'] <= pd.Timestamp('2018-07-01'))]
```

```python
# lockdown 2019
df[(df['occurred_on_date'] >= pd.Timestamp('2019-03-15')) & (df['occurred_on_date'] <= pd.Timestamp('2019-07-01'))]
```

```python
# lockdown 2021
df[(df['occurred_on_date'] >= pd.Timestamp('2021-01-01')) & (df['occurred_on_date'] <= pd.Timestamp('2021-12-31'))]
```
