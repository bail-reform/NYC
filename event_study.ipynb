```python
import pandas as pd

# Load the dataset
data = pd.read_csv('/Users/yujinlee/Google_Drive/NYU/Fall_2023/Capstone/CrimeData/clean/nyc_la_stata.csv')
data = data.loc[:, ~data.columns.str.contains('^Unnamed')]

# Display the first few rows of the dataframe to understand its structure
data.head()
```

```python
from datetime import datetime

# Convert the week column to datetime format
data['week'] = pd.to_datetime(data['week'])

# Define the event date for the bail reform
event_date = datetime(2020, 1, 1)

# Calculate the number of weeks relative to the event date
data['weeks_from_event'] = (data['week'] - event_date).dt.days / 7
data['weeks_from_event'] = data['weeks_from_event'].round()

# Create dummy variables for the three weeks before and three weeks after the event
# Note: The fourth week before the event is the reference period (no dummy needed)
for i in range(-3, 4):
    if i != -4:  # Skip the reference week
        data[f'event_week{i}'] = (data['weeks_from_event'] == i).astype(int)
```

```python
data[['week', 'weeks_from_event', 'event_week-3', 'event_week-2', 'event_week-1', 'event_week1', 'event_week2', 'event_week3']].head()
```

```python
data[data['event_week0'] != 0]
```

```python
data.to_csv('/Users/yujinlee/Google_Drive/NYU/Fall_2023/Capstone/CrimeData/clean/nyc_la_eventstudy_stata.csv', index=False)
```
